Outlier Detection and Removal:





```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.boxplot(data['Age'].dropna())
plt.title('Age')

plt.subplot(1, 3, 2)
plt.boxplot(data['Experience_years'].dropna())
plt.title('Years of Experience')

plt.subplot(1, 3, 3)
plt.boxplot(data['Salary'].dropna())
plt.title('Salary')

plt.tight_layout()
plt.show()
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

for col in ['Age', 'Experience_years', 'Salary']:
    data = remove_outliers(data, col)
```

Label Encoding:

```python
from sklearn.preprocessing import LabelEncoder
Label_Encoder = LabelEncoder()
```

```python
gender_encoder = LabelEncoder()
degree_encoder = LabelEncoder()
job_encoder = LabelEncoder()
```

```python
data1['Gender'] = gender_encoder.fit_transform(data1['Gender'])
data1['Degree'] = degree_encoder.fit_transform(data1['Degree'])
data1['Job_Title'] = job_encoder.fit_transform(data1['Job_Title'])
```

```python
x=data1.drop(columns=['Salary'])
y=data1['Salary']
x
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "AdaBoost": AdaBoostClassifier(),
}
```

```python
results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.bar(results.keys(), results.values(), color='skyblue')
plt.ylabel('Accuracy Score')
plt.title('Model Comparison')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
results = {}

for name, model in models.items():
    model.fit(x_train, y_train)
    preds = model.predict(x_test)
    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print(f"{name}: {acc:.4f}")
```

```python
selected_model_S = max(results, key=results.get)
selected_model = models[selected_model_S]
print(f"\n Selected model: {selected_model_S} with accuracy {results[selected_model_S]:.4f}")
```

```python
import joblib
joblib.dump(selected_model, "selected_model.pkl")
print("Saved model as selected_model.pkl")
```

```python
import gradio as gr
import numpy as np
import joblib

model = joblib.load('selected_model.pkl')
gender_encoder = joblib.load('gender_encoder.joblib')
degree_encoder = joblib.load('degree_encoder.joblib')
job_encoder = joblib.load('job_encoder.joblib')

def predict_salary(age, gender, degree, job_title, experience):
    try:
        gender_encoded = gender_encoder.transform([gender])[0]
        degree_encoded = degree_encoder.transform([degree])[0]
        job_encoded = job_encoder.transform([job_title])[0]

        input_data = np.array([[gender_encoded, degree_encoded, job_encoded, age, experience]])

        salary = model.predict(input_data)[0]
        return f"Predicted Salary: ₹{salary:,.2f}"

    except Exception as e:
        return f"Error: {str(e)}"

demo = gr.Interface(
    fn=predict_salary,
    inputs=[
        gr.Number(label="Age"),
        gr.Radio(gender_encoder.classes_.tolist(), label="Gender"),
        gr.Radio(degree_encoder.classes_.tolist(), label="Degree"),
        gr.Dropdown(job_encoder.classes_.tolist(), label="Job Title"),
        gr.Number(label="Experience (Years)"),
    ],
    outputs="text",
    title="Employee Salary Prediction App",
    description="Enter employee details to predict salary:"
)

demo.launch(share=True)
```
